When you load a GeoJSON file using the json library, you get a dict that contains an entry features, which contains the list of features. Each feature in turn consists of a dict, which, among other things, contains an entry geometry. The geometry is a dict containing the entries type and coordinates. 

```python
import pandas as pd
```

```python
# to traverse GeoJSON file

import json

with open('./datasets/master-plan-2019-sdcp-unesco-heritage-site-layer/master-plan-2019-sdcp-unesco-heritage-site-layer-geojson.geojson') as f:
    data = json.load(f)

for feature in data['features']:
    print(feature['geometry']['type'])
    print(feature['geometry']['coordinates'])
```

```python
print(data.values())
```

```python
pd.DataFrame.from_dict(data)
```
